# **WEB SCRAPING**

 Web scraping refers to the extraction of data from a website. This information is collected and then exported into a format that is more useful for the user. Be it a spreadsheet or an API.Most web scrapers will output data to a CSV or Excel spreadsheet, while more advanced scrapers will support other formats such as JSON which can be used for an API.

 Ideally, the user will go through the process of selecting the specific data they want from the page. For example, you might want to scrape an Amazon product page for prices and models but are not necessarily interested in product reviews.
 Lastly, the web scraper will output all the data that has been collected into a format that is more useful to the user.

# Importing Modules

```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
```

# Scraping the Data from FLIPKART

```python
url="https://www.flipkart.com/search?q=television"
```

# Request pass to the flipkart page

```python
r=requests.get(url)
```

# To store the web page data in one place

```python
soup=BeautifulSoup(r.content,"html.parser")
```

# Scrap the Titles Data

```python
titles=soup.find_all('div',class_='_KzDlHZ')
```

# Scrap the Ratings Data


```python
ratings=soup.find_all('div',class_='_5OesEi')
```

# Scrap the Reviews Data

```python
reviews=soup.find_all('span',class_="_Wphh3N")
```

# Scrap the Prices  Data

```python
prices=soup.find_all('div',class_='_hl05eU')
```

# Storing the all scraping data into the List

```python
tvtitles=[]
tvratings=[]
tvreviews=[]
tvprices=[]
```

# Storing the all collected data into ZIP Format

```python
for titles,ratings,reviews,prices in zip(titles,ratings,reviews,prices):
  tvtitles.append(titles.text)
  tvratings.append(ratings.text)
  tvreviews.append(reviews.text)
  tvprices.append(prices.text)

```

collecting the all stored data and these data form into the zip format and these data are change to the text format

```python
print(tvtitles)
print(tvratings)
print(tvreviews)
print(tvprices)
```

# Saving data in CSV file

```python
data={'Titles:tvtitles,Ratings:tvratings,Reviews:tvreviews,Prices:tvprices'}
```

using dictionary

```python
print(data)
```

# Using Pandas

```python
model=pd.DataFrame(data=data)
```

Using pandas to create the data form into rows and columns

```python
model.to_csv('television.csv')
```
